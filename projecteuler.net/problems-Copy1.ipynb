```python
def make_port_ranges(cli_input):
    
    max_port = 65355
    allowed = cli_input.split(' ')
    int_allowed_ports = []
    
    for i in allowed:
        int_allowed_ports.append(int(i))
    
    sorted_ports = sorted(int_allowed_ports)
    j = 0
    pairs = []
    for i in sorted_ports:
        if j == 0:
            start = 1
        else:
            start = sorted_ports[j-1] + 1
        end = i - 1
        j += 1
        formatted = str(start) + ':' + str(end)
        pairs.append(formatted)
        #add the last port range here
        if i == sorted_ports[-1]:
            start = i + 1
            end = max_port
            formatted = str(start) + ':' + str(end)
            pairs.append(formatted)
    print(pairs)
    return ' '.join(pairs)
cli_input = '2280 50 80 22'
make_port_ranges(cli_input)
```
